# delete operator in python

**del** is delete operator in python

**It is used to remove the given object reference from the memory.**

Now given object referece cannot be accessed

Ex.

num=999 # num is object reference to 999

del num # delete object reference num

print(num) # This given NameError




# del when used on a list / iterable (list, set, dictionary)

## case 1 : delete the complete object

del will remove the refrence which is given to it

Ex.

l1=[ 1,2,3 ]

del l1 # remove reference l1 , so complete list is deleted.

s1= {8,9,5}

del s1 # deletes reference s1, so complete set is deleted

d1 = {'k1':'v1','k2':'v2'}

del d1 # deletes reference d1, so complete dictionary is deleted





## case 2 : delete an element inside object (like list / set / dictionary)

l1=[34,32,21]

del l1[0] # deletes reference to first element in list (34). Because of this len of list will be reduced by 1

print(l1) # [32, 21]





# How to delete a list in python ?

## Recommended way

l1 = [1,2,3]

del l1

#### 'del l1' deleted reference l1. Then python garbage collector will take care to remove remaining references and objects. Garbage collector uses reference counting technique to decide which object to remove from heap

# What happens if we delete elements of list in a loop ?


## Wrong way

#### In following code ... copy of the element is removed, original list remains same

```python
# wrong way of deleting an element
l1=[23,56,78,90,999]
for elem in l1:
    print("deleting element =", elem)
    del elem
    print(len(l1),l1)
```

#### In following code original list is modified ... but it gives index error and all elements will not be deleted

```python
# delete elements inside the loop
l1=[23,56,78,90,999]
for idx in range(len(l1)):
    print("deleting by idx=", idx)
    print(" element=",l1[idx])
    del l1[idx]
    print(len(l1),l1)
    if (idx == 3):
        break
```

# Right way to delete elements inside the loop

There can be few ways to delete all elements.

Here listing 3 of them

### Method 1 Using del and shallow copy

```python
# Right way to delete elements inside the loop
l1=[23,56,78,90,999]
l2=l1.copy()
for idx in range(len(l1)):
    print("deleting by idx=", idx)
    print(" element=",l2[idx])
    l1_idx = l1.index(l2[idx])
    del l1[l1_idx]
    print(len(l1),l1)
```

#### Method 2 : Using del ... delete first element every time in a loop

```python
# Right way to delete elements inside the loop
l1=[23,56,78,90,999]
for idx in range(len(l1)):
    print("deleting first element every time")
    print("deleted element=",l1[0])
    del l1[0]
```

#### Method 3: Using pop function .. delete last element every time in a loop

```python
# Right way to delete elements inside the loop
l1=[23,56,78,90,999]
for idx in range(len(l1)):
    print("deleting by using pop")
    elem = l1.pop()
    print("deleted element=",elem)
```

```python
import re
regex1="i*"
regex2="si*"
S1="iacsd"
S2="sii"
S3="abc"
```

```python
re.findall(regex1,S1)
```

```python
re.findall(regex1,S2)
```

```python
re.findall(regex1,S3)
```

```python
re.findall(regex2,S1)#si*
```

```python
re.findall(regex2,S2)
```

```python
re.findall(regex2,S3)
```

# {m,n} Example

```python
regex="[0-9]{2,4}"
S1="456"
S2="9BN"
S3="67HJK890780"
```

```python
re.findall(regex,S1)
```

```python
re.findall(regex,S2)
```

```python
re.findall(regex,S3)
```

# Use Search

```python
regex="[0-9]{2,4}"
S1="456"
S2="9BN"
S3="67HJK890780"
```

```python
re.search(regex,S1)
```

```python
m = re.search(regex,S1)
m[0]
```

```python
S1[0:3]
```

```python
m.span()
```

```python
m = re.search(regex,S3)
m[0]
```
